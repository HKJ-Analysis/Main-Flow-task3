# Task - 3: Create a bar chart and a line chart using Matplotlib to visualize data from a Pandas DataFrame. Customize the charts with labels, titles, and legends.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("householdtask3.csv")
data
```

```python
# Line Chart with year and own

plt.plot(data['year'])
plt.plot(data['own'])

# x and y co-ordinates labels.
plt.xlabel('Year')
plt.ylabel('Own')

# Title of Chart
plt.title('Line Chart')

# To Display chart on screen.
plt.show()
```

```python
# Bar Chart with year and own.

plt.bar(data['year'],data['own'])

# x and y co-ordinates label.
plt.xlabel('Year')
plt.ylabel('Own')

# Title of Chart
plt.title('Bar Chart')

# To Display chart on screen
plt.show()
```

```python
# Scatter chart with year and Own.

plt.scatter(data['year'],data['own'],color = 'red')

# x and y co-ordinates label.
plt.xlabel('Year')
plt.ylabel('Own')

# Titile of chart
plt.title('Scatter Chart')

# To Display chart on screen.
plt.show()
```

```python
# Histogram chart with year and own.

plt.hist(data['income'],color='#005C78')

# Title of Chart
plt.title('Histogram Chart')

# To Display chart oon screen.
plt.show()
```
